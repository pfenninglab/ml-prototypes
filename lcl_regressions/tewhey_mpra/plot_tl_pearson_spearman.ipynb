```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
rcParams['svg.fonttype'] = 'none'
sns.set(font_scale = 2)
sns.set_style("white")
rcParams['svg.fonttype'] = 'none'

import pandas as pd
```

```python
pearsons = {'TL0': [0.39, 'pearson'],
 'TL1': [0.37, 'pearson'],
 'TL2': [0.31, 'pearson'],
 'TL3': [0.41, 'pearson'],
 'DNase': [0.3, 'pearson'],
 'MPRA': [0.297, 'pearson']
}
```

```python
spearmans = {'TL0': [0.38,'spearman'],
 'TL1': [0.34, 'spearman'],
 'TL2': [0.28, 'spearman'],
 'TL3': [0.37, 'spearman'],
 'DNase': [0.22, 'spearman'],
 'MPRA': [0.227, 'spearman']
}
```

```python
df_pearson = pd.DataFrame.from_dict(pearsons, orient='index')
df_spearman = pd.DataFrame.from_dict(spearmans, orient='index')
```

```python
df = pd.concat([df_pearson, df_spearman], axis=0)
```

```python
df.columns = ['corr', 'corr_type']
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df,
            x=df.index,
            y=df['corr'],
            hue=df['corr_type']
           )
plt.savefig('plots/pearson_spearman_test_set_tl_models.svg')
```
