```python
def revcomp(inseq):
    mapDict = {"A": "T", "T": "A", "C": "G", "G": "C", "N": "N"}
    outseq = ""
    for nuc in inseq:
        nucUpper = nuc.upper()
        outseq = mapDict[nucUpper] + outseq
    return outseq
```

```python
def oneHotEncodeSequence(sequence):
    oneHotDimension = (len(sequence), 4)
    dnaAlphabet = {"A":0, "G":1, "C":2, "T":3}
    one_hot_encoded_sequence = np.zeros(oneHotDimension, dtype=np.int)
    for i, nucleotide in enumerate(sequence):
        if nucleotide.upper() in dnaAlphabet:
            index = dnaAlphabet[nucleotide.upper()]
            one_hot_encoded_sequence[i][index] = 1
    return one_hot_encoded_sequence
```

```python
leftWindow = 499
rightWindow = 500
finalReferenceSequences = []
finalAlternateSequences = []
skews = []
skews1 = []
for index, snp in snp_info.iterrows():
    chrom = snp["CHR"]
    start = snp["START"]
    end = snp["END"]
    rsid = index
    orientation = snp["Direction"]
    
    a1 = snp["A1"]
    a2 = snp["A2"]
    if chrom==None or a1==None or a2==None or not rsid.startswith("rs"):
        continue
    if len(a2)>1:
        continue
    hg38_a1 = genomeObject[chrom][start].upper()
    hg38_a2 = a2
    if not hg38_a1.upper() == a1:
        if hg38_a1.upper() != a2:
            continue
        assert(hg38_a1.upper()==a2)
        hg38_a2 = a1
    ref_sequence = genomeObject[chrom][start - leftWindow:start] + hg38_a1 + genomeObject[chrom][end: end+rightWindow]
    alt_sequence = genomeObject[chrom][start-leftWindow:start] + hg38_a2 + genomeObject[chrom][end: end+rightWindow]
    
    if orientation == "neg":
        ref_sequence = revcomp(ref_sequence)
        alt_sequence = revcomp(alt_sequence)
    assert(len(ref_sequence)==1000)
    if len(alt_sequence)!=1000:
        print snp
    assert(len(alt_sequence)==1000)
    encodedReferenceSequence = oneHotEncodeSequence(ref_sequence)
    encodedAlternateSequence = oneHotEncodeSequence(alt_sequence)
    finalReferenceSequences.append(encodedReferenceSequence)
    finalAlternateSequences.append(encodedAlternateSequence)
    skews.append(snp["LogSkew.Comb"])
    skews1.append(snp["LogSkew.12878"])

finalReferenceSequences = np.stack(finalReferenceSequences, axis=0)
finalAlternateSequences = np.stack(finalAlternateSequences, axis=0)

print finalReferenceSequences.shape
print finalAlternateSequences.shape
np.save("/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_lcl/tewhey_reference_sequences.npy", finalReferenceSequences)
np.save("/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_lcl/tewhey_alternate_sequences.npy", finalAlternateSequences)
```
