```python
import pandas as pd
import numpy as np
from keras.models import Sequential, load_model
from scipy.stats import pearsonr, spearmanr
import matplotlib.pyplot as plt
from matplotlib import rcParams
from regression_model import pearson_correlation
rcParams.update({'figure.autolayout': True})
rcParams['svg.fonttype'] = 'none'
rcParams['font.size']=10
```

```python
import keras.metrics
keras.metrics.pearson_correlation = pearson_correlation
```

```python
model_2 = load_model('/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_lcl/model_2.hdf5')
model_6 = load_model('/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_lcl/model_6.hdf5')
model_7 = load_model('/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_lcl/model_7.hdf5')
model_8 = load_model('/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_lcl/model_8.hdf5')
model_9 = load_model('/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_lcl/model_9.hdf5')
```

```python
X_train = np.load('/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_lcl/chrs_training_summit_centered_training_set_X.npy')
Y_train = np.load('/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_lcl/chrs_training_summit_centered_training_set_Y.npy')
X_valid = np.load('/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_lcl/chr4_summit_centered_validation_set_X.npy')
Y_valid = np.load('/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_lcl/chr4_summit_centered_validation_set_Y.npy')
```

```python
model_2_pred_train = model_2.predict(X_train)
model_2_pred_valid = model_2.predict(X_valid)
```

```python
f, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

plt.title("Model 2 Performance")

ax1.scatter(Y_train, model_2_pred_train)
ax1.set_xlabel("True Fold Change (training)")
ax1.set_ylabel("Model 2 Predicted Fold Change (training)")
ax2.scatter(Y_valid, model_2_pred_valid)
ax2.set_xlabel("True Fold Change (validation)")
ax2.set_ylabel("Model 2 Predicted Fold Change (validation)")
plt.show()
print("Training:",   pearsonr(Y_train, model_2_pred_train))
print("Training:",   spearmanr(Y_train, model_2_pred_train))

print("Validation:",   pearsonr(Y_valid, model_2_pred_valid))
print("Validation:",   spearmanr(Y_valid, model_2_pred_valid))
```
