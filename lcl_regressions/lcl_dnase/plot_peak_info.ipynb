```python
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
rcParams['svg.fonttype'] = 'none'
rcParams['font.size']=10
```

```python
gm_conservative_narrowpeak_file = '/projects/pfenninggroup/machineLearningForComputationalBiology/addiction_gwas_enrichment/lcl_variant_prediction/lcl_human/DNase-seq_ENCODE_GM12878/peak/idr_reproducibility/idr.conservative_peak.narrowPeak.gz'


gm_conservative_peaks_info = pd.read_csv(gm_conservative_narrowpeak_file, compression='gzip', 
                                 names=["CHR", "START", "END", "NAME", "SCORE", "STRAND", "SIGNAL", "P", "Q", "SUMMIT"], sep='\t')
```

```python
fig, ax = plt.subplots()

parts = ax.violinplot(dataset=[gm_conservative_peaks_info.SIGNAL])

parts['bodies'][0].set_facecolor('r')
parts['bodies'][0].set_edgecolor('black')
parts['bodies'][0].set_alpha(1)

plt.title("Fold Change Signal")
plt.ylabel("Fold Change Signal")
plt.boxplot(x=[gm_conservative_peaks_info.SIGNAL])
plt.show()
```

```python
fig, ax = plt.subplots()

parts = ax.violinplot(dataset=[np.log(gm_conservative_peaks_info.SIGNAL)])

parts['bodies'][0].set_facecolor('r')
parts['bodies'][0].set_edgecolor('black')
parts['bodies'][0].set_alpha(1)

plt.title("log(Fold Change Signal)")
plt.ylabel("log(Fold Change Signal)")
plt.boxplot(x=[np.log(gm_conservative_peaks_info.SIGNAL)])
plt.show()
```

```python
fig, ax = plt.subplots()

parts = ax.violinplot(dataset=[gm_conservative_peaks_info.END-gm_conservative_peaks_info.START])

parts['bodies'][0].set_facecolor('b')
parts['bodies'][0].set_edgecolor('black')
parts['bodies'][0].set_alpha(1)

plt.title("Peak Length")
plt.ylabel("Peak Length")
plt.boxplot(x=[gm_conservative_peaks_info.END-gm_conservative_peaks_info.START])
plt.show()
```
